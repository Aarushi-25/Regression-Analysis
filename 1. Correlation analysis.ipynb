```python
#Step 1: Importing the libraries.
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Step 2: Finding the Correlation between two variables.
y = pd.Series([1, 2, 3, 4, 3, 5, 4])
x = pd.Series([1, 2, 3, 4, 5, 6, 7])
correlation = y.corr(x)
correlation
```

```python
#Step 3: Plotting the graph. Here we are using scatter plots. A scatter plot is a diagram where each value in the data set is represented by a dot. Also, it shows a relationship between two variables.
# plotting the data
plt.scatter(x, y)
 
# This will fit the best line into the graph
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))
         (np.unique(x)), color='red')


#Adding title and labels in the graph
plt.title('Correlation')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
import seaborn as sns
# Step 2: Creating a DataFrame with your data.
data = pd.DataFrame({'x': [1, 2, 3, 4, 5, 6, 7], 'y': [1, 2, 3, 4, 3, 5, 4]})
# Step 3: Calculate the correlation matrix.
correlation_matrix = data.corr()
correlation_matrix
```

```python
# Step 4: Plotting the correlation matrix as a heatmap.
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#calculating from the formula
import numpy as np
x = np.array([1,3,5,7,8,9, 10, 15])
y = np.array([10, 20, 30, 40, 50, 60, 70, 80])
#Define the correlation by applying the above formula:
def Pearson_correlation(X,Y):
    if len(X)==len(Y):
        Sum_xy = sum((X-X.mean())*(Y-Y.mean()))
        Sum_x_squared = sum((X-X.mean())**2)
        Sum_y_squared = sum((Y-Y.mean())**2)       
        corr = Sum_xy / np.sqrt(Sum_x_squared * Sum_y_squared)
    return corr           
print(Pearson_correlation(x,y))    
print(Pearson_correlation(x,x))
```
