```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def forecast_accuracy(forecast, actual):
    # Calculate forecast error
    error = forecast - actual
    
    # Calculate forecast bias
    bias = np.mean(error)
    
    # Calculate forecast accuracy metrics
    mae = np.mean(np.abs(error))  # Mean Absolute Error
    mape = np.mean(np.abs(error) / actual)  # Mean Absolute Percentage Error
    mse = np.mean(error**2)  # Mean Squared Error
    rmse = np.sqrt(mse)  # Root Mean Squared Error
    
    return {'MAE': mae, 'MAPE': mape, 'MSE': mse, 'RMSE': rmse, 'Bias': bias}
data = {'Period': [1, 2, 3, 4, 5, 6],
        'Actual': [10, 12, 8, 15, 11, 13],
        'Forecast': [9, 10, 8, 13, 12, 11]}
df = pd.DataFrame(data)

# Add error column to the dataframe
df['Error'] = df['Forecast'] - df['Actual']

# Calculate forecast accuracy, bias, and error
accuracy_results = forecast_accuracy(df['Forecast'], df['Actual'])

# Print the dataframe
print("Time Series Data:")
print(df)

# Print the results
print("\nForecast Accuracy Metrics:")
for metric, value in accuracy_results.items():
    print(f"{metric}: {value}")

# Plot actual vs forecast
plt.plot(df['Period'], df['Actual'], label='Actual')
plt.plot(df['Period'], df['Forecast'], label='Forecast')
plt.xlabel('Period')
plt.ylabel('Demand')
plt.title('Actual vs Forecast')
plt.legend()
plt.show()
```
